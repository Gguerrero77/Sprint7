## Ejercicio 1
Calculadora del índice de masa corporal

Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos.

El cálculo del IMC
Para calcular el IMC es necesario dividir el peso en kg de un paciente entre el cuadrado de su altura en metros. Con esto obtendremos un valor que se situará en uno de los siguientes rangos:

Bajo peso: menos de 18,5.
Peso normal: 18,5-24,9.
Sobrepeso: 25-29,9.
Obesidad: más de 30.
De este modo se puede catalogar la situación corporal del paciente con un sencillo cálculo, lo que permite identificar un posible problema de salud para empezar el tratamiento. Esta fórmula ha sido muy utilizada en consultas de endocrinología y de cardiología durante muchos años, ya que se utilizaba como predictor del riesgo cardiovascular.

```python
def IMCfunc ():   
    x = float(input("Indica tu peso en Kg.: "))
    print("Tu peso: ", x, "Kg.")
    y = float(input("Indica tu estatura en mts.: "))
    print("Tu estatura", y, "mts.")
    IMC = x/y**2
    print(f"IMC = {round(IMC, 2)}")
    if IMC <18.5:
        print("Bajo peso")
    elif IMC <= 24.9:
        print("Peso normal")
    elif IMC <= 29.9:
        print("Sobrepeso")
    else: 
        print("Obesidad")
```

## Ejercicio 2
Conversor de temperaturas.

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re). Selecciona al menos 2 y construye el conversor.

Para el ejercicio he seleccionado Celsius, Fahrenheit y Kelvin.

```python
def caf(celsius):
    return round(((celsius * 9/5) + 32), 2)

def cak(celsius):
    return round((celsius + 273.15), 2)

def fac(fahrenheit):
    return round(((fahrenheit - 32) * 5/9), 2)

def fak(fahrenheit):
    return round(((fahrenheit - 32) * 5/9 + 273.15), 2)

def kac(kelvin):
    return round((kelvin - 273.15), 2)

def kaf(kelvin):
    return round(((kelvin - 273.15) * 9/5 + 32), 2)

# Diccionarios para conversiones directas
conversions = {
    '1': {
        '2': caf,
        '3': cak
    },
    '2': {
        '1': fac,
        '3': fak
    },
    '3': {
        '1': kac,
        '2': kaf
    }
}

# Diccionario para nombres de las unidades
unit_names = {
    '1': 'ºCelsius',
    '2': 'ºFahrenheit',
    '3': 'ºKelvin'
}

def Conversor():
    x = input(f"Selecciona la unidad de entrada {unit_names}")
    y = input(f"Selecciona la unidad de salida {unit_names}")
    z = float(input("Ingresa temperatura:"))

    if x not in conversions or y not in unit_names:
        print("¡Error!, seleccione 1, 2 o 3")
        Conversor()
    elif x == y:
        print(f"{z} {unit_names[x]}")
    else:
        conversion_func = conversions[x].get(y)
        converted_value = conversion_func(z)
        print(f"{z} {unit_names[x]} = {converted_value} {unit_names[y]}")
        
```

## Ejercicio 3
Contador de palabras de un texto.

Escribe una función que dado un texto, muestre las veces que aparece cada palabra.

En esta funcion los archivos se cargan desde la carpeta donde se almacena el script.

Archivo: tu_me_quieres_blanca.txt

```python
import re #libreria que permite eliminar signos de puntuacion y caracteres especiales en el texto https://docs.python.org/es/3/library/re.html
from itertools import islice #libreria que permite imprimir el numero n de lineas del archivo de texto dado

def Contador():
    
    x = (input("Ingresa el nombre del archivo: "))
    print(x)
    n = 7
    with open(x, "r", encoding="utf-8") as archivo: # Abre el archivo en modo lectura
         
        for texto in islice(archivo, n): #recorta el texto pra imprimir solo las lineas que se indican en la variable n.
            print(texto)
            
    with open(x, "r", encoding="utf-8") as archivo: # Abre el archivo en modo lectura nuevamente para procesar el texto completo.        
            LibroPrueba = archivo.read() # Muestra el contenido del archivo
   
    LibroPrueba = re.sub(r'[^\w\s]', '', LibroPrueba)  
    ''' Eliminar signos de puntuación y caracteres especiales: Utilizo re.sub(r'[^\w\s]', '', LibroPrueba) 
    para reemplazar cualquier carácter que no sea una palabra (\w) o un espacio (\s) con una cadena vacía
    de esta manera se evitan errores como contar "hola:" como una palabra distinta de "hola" o contar un punto como si fuera una palabra '''

    palabras = LibroPrueba.split() # Divide el contenido en palabras

    CountPalabras = len(palabras) # Cuenta el número de palabras
    print(f"El número de palabras en el archivo es: {CountPalabras}")
    
    # Cuenta las ocurrencias de cada palabra
    ocurrencias = {}
    for palabra in palabras:
        palabra = palabra.lower()  # Convertir a minúsculas para contar sin distinción de mayúsculas/minúsculas
        if palabra in ocurrencias:
            ocurrencias[palabra] += 1
        else:
            ocurrencias[palabra] = 1

    # Muestra el conteo de cada palabra
    for palabra, count in ocurrencias.items():
        print(f"La palabra '{palabra}' se repite {count} veces.")
```

```python
Contador()
```

## Ejercicio 4

Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.

```python
def invertir_diccionario(diccionario):
            
    valores = list(diccionario.values())
    # Verificar si todos los valores son únicos
    
    if len(valores) != len(set(valores)):
        print("Advertencia: Las claves en el diccionario invertido no seran únicas.")
        return None
   
        # Invertir el diccionario
    diccionario_invertido = {valores: clave for clave, valores in diccionario.items()}
    
    print(diccionario_invertido)

#La salida de esta funcion seria: Diccionario invertido: {'Ford': 'brand', 'Mustang': 'model', 'azul': 'year'} 
#La clave 'year' con el valor 2020 se omite porque fue sobrescrita en el diccionario original por 'year': 'azul'. 
#Python no permite claves duplicadas en un diccionario, y la última aparición de una clave sobrescribe las anteriores.
```

```python
# Ejemplo en este caso los valores del diccionario original estan duplicados se emite el mensaje de error y no se procesa
encuesta_antigua = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 2020,
    "color": 2020
}

invertir_diccionario(encuesta_antigua)
```

```python
# Ejemplo en este caso los valores del diccionario original no estan duplicados se omite el mensaje de error y se procesa
encuesta_antigua = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 2020,
    "color": "azul"
}
invertir_diccionario(encuesta_antigua)
```

## Ejercicio 1
Diccionario inverso con duplicados

Continuando con el ejercicio 4 del nivel 1: al cliente se olvidó comentar un detalle y resulta que los valores en el diccionario original pueden duplicarse y más, por lo que las claves intercambiadas pueden tener duplicados. En este caso, en el ejercicio anterior imprimías un mensaje de advertencia, ahora, los valores del diccionario resultante tendrán que almacenarse como una lista. Ten en cuenta que si es un valor único no debe ser una lista.

```python
def invertir_diccionario(diccionario):
    valores = list(diccionario.values())
    # Verificar si todos los valores son únicos
    
    if len(valores) != len(set(valores)):
        print("Advertencia: Las claves en el diccionario invertido no serían únicas, los valores se agregan en lista en una misma clave.")
        
    # Nuevo diccionario invertido
    diccionario_invertido = {}

    for clave, valor in diccionario.items():
        if valor in diccionario_invertido:
            # Si el valor ya está en el nuevo diccionario, agrega la clave a la lista existente
            if isinstance(diccionario_invertido[valor], list):
                diccionario_invertido[valor].append(clave)
            else:
                diccionario_invertido[valor] = [diccionario_invertido[valor], clave]
        else:
            # Si el valor no está en el nuevo diccionario, añade la clave correspondiente
            diccionario_invertido[valor] = clave

    print(diccionario_invertido)
```

```python
# Ejemplo
encuesta_antigua = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 2020,
    "color": 2020
}

invertir_diccionario(encuesta_antigua)
```

## Ejercicio 2
Conversión de tipos de datos

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number' ), [1,2,'3','3.4']]

```python
def es_convertible(elemento):   #esta funcion sirve para comprobar si cada elemento de la lista es convertible
    try:                        #(sub-elemntos tambien en el caso de estudio ya que dentro de nuestra lista hay listas y tuplas)
        float(elemento)         #se trata de una comprobación que arroja un resultado logico True o False
        return True
    except (ValueError, TypeError):
        return False

def Flotar_lista(lista):      # Esta funcion llama a la funcion previa es_convertible su finalidad es, tras comprobar 
    lista_convertibles = []     #la convertibilidad de cada elemento y sub-elemento agregarlos a una lista segun sean o no convertibles.
    lista_no_convertibles = []

    for elemento in lista:
        if isinstance(elemento, (list, tuple)):
            for sub_elemento in elemento:
                if es_convertible(sub_elemento):
                    lista_convertibles.append(float(sub_elemento))
                else:
                    lista_no_convertibles.append(sub_elemento)
        else:
            if es_convertible(elemento):
                lista_convertibles.append(float(elemento))
            else:
                lista_no_convertibles.append(elemento)
                
    print("elementos convertidos", lista_convertibles)
    print("elementos no convertidos", lista_no_convertibles)
```

```python
# Lista de ejemplo proporcionada
lista_cliente = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

Flotar_lista(lista_cliente)
```

## Ejercicio 1
Contador y ordenador de palabras de un texto.

El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a la Z. Por ejemplo, para el archivo "tu_me_quieres_blanca.txt".

```python
import re
import json
from collections import defaultdict

def Contador_ordenado():
    x = input("Ingresa el nombre del archivo: ")
    print(x)
    with open(x, "r", encoding="utf-8") as archivo:
        LibroPrueba = archivo.read()

    LibroPrueba = re.sub(r'[^\w\s]', '', LibroPrueba)
    palabras = LibroPrueba.split()

    CountPalabras = len(palabras)
    print(f"El número de palabras en el archivo es: {CountPalabras}")

    ocurrencias = {}
    for palabra in palabras:
        palabra = palabra.lower()  # Convertir a minúsculas para contar sin distinción de mayúsculas/minúsculas
        if palabra in ocurrencias:
            ocurrencias[palabra] += 1
        else:
            ocurrencias[palabra] = 1

    # Clasificar palabras por la primera letra
    clasificado_por_letra = defaultdict(dict)
    for palabra, ocurrencia in ocurrencias.items():
        primera_letra = palabra[0]
        clasificado_por_letra[primera_letra][palabra] = ocurrencia

    # Ordenar las palabras dentro de cada letra
    resultado_ordenado = {letra: dict(sorted(clasificado_por_letra[letra].items())) 
                          for letra in sorted(clasificado_por_letra.keys())}
    
    formato_salida = json.dumps(resultado_ordenado, indent=2, ensure_ascii=False) 
    ''' la función da formato a la salida con indentación de los items en el diccionario y para que se muestren
    caracteres especiales como la ñ o letras acentuadas se agrego el parametro (ensure_ascii=False)'''

    print(formato_salida)
```
